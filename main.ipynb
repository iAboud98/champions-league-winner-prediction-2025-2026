```python
teams = pd.concat([
    df[["hometeam", "homeelo"]].rename(columns={"hometeam":"team","homeelo":"elo"}),
    df[["awayteam", "awayelo"]].rename(columns={"awayteam":"team","awayelo":"elo"})
])

team_mean = teams.groupby("team")["elo"].mean()
global_mean = teams["elo"].mean()
```

```python
df["homeelo"] = df["homeelo"].fillna(
    df["hometeam"].map(team_mean)
)

df["homeelo"] = df["homeelo"].fillna(global_mean)

df["awayelo"] = df["awayelo"].fillna(
    df["awayteam"].map(team_mean)
)

df["awayelo"] = df["awayelo"].fillna(global_mean)
```

```python
features = [
    ("homeyellow",  "hometeam"),
    ("awayyellow",  "awayteam"),
    ("homered",     "hometeam"),
    ("awayred",     "awayteam"),
    ("homeshots",   "hometeam"),
    ("awayshots",   "awayteam"),
    ("hometarget",  "hometeam"),
    ("awaytarget",  "awayteam"),
    ("homefouls",   "hometeam"),
    ("awayfouls",   "awayteam"),
    ("homecorners", "hometeam"),
    ("awaycorners", "awayteam")
]

for col, team_col in features:
    team_median = df.groupby(team_col)[col].median()
    global_median = df[col].median()
    
    df[col] = df[col].fillna(df[team_col].map(team_median))
    df[col] = df[col].fillna(global_median)
```

```python
df["season"] = df["matchdate"].dt.year.where(
    df["matchdate"].dt.month >= 7,
    df["matchdate"].dt.year - 1
)

cols = [
    "over25",
    "under25",
    "maxover25",
    "maxunder25",
    "handisize",
    "handihome",
    "handiaway"
]

for col in cols:
    
    group_median = df.groupby(["division","season"])[col].transform("median")
    
    df[col] = df[col].fillna(group_median)
    df[col] = df[col].fillna(df[col].median())
```

```python
numeric_cols = df.select_dtypes(include=['float64','int64','int32']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].median()
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 2.5 * IQR
    upper = Q3 + 2.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    
    if len(outliers) > 0:
        print(f"Feature: {col}")
        print(f"Q1: {Q1}")
        print(f"Q2 (Median): {Q2}")
        print(f"Outliers: {outliers.sort_values(ascending=False).tolist()}")
        print("-" * 30)
```

```python
cols_to_clip = ['oddwin', 'oddloss', 'maxwin', 'maxloss']

for col in cols_to_clip:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    upper = Q3 + 2.5 * IQR
    
    df[col] = df[col].clip(upper=upper)

print("Clipped values to upper bound for:", cols_to_clip)
```

```python
# Create difference features
df['elodiff'] = df['homeelo'] - df['awayelo']
df['form5diff'] = df['form5home'] - df['form5away']
df['ftgoalsdiff'] = df['fthome'] - df['ftaway'] 
df['htgoalsdiff'] = df['hthome'] - df['htaway']
df['shotsdiff'] = df['homeshots'] - df['awayshots'] 
df['targetdiff'] = df['hometarget'] - df['awaytarget']
df['foulsdiff'] = df['homefouls'] - df['awayfouls'] 
df['cornersdiff'] = df['homecorners'] - df['awaycorners']
df['reddiff'] = df['homered'] - df['awayred']
df['yellowdiff'] = df['homeyellow'] - df['awayyellow']

# Drop original columns after creating differences
cols_to_drop = [
    'homeelo', 'awayelo',
    'form5home', 'form5away',
    'fthome', 'ftaway',
    'hthome', 'htaway',
    'homeshots', 'awayshots',
    'hometarget', 'awaytarget',
    'homefouls', 'awayfouls',
    'homecorners', 'awaycorners',
    'homered', 'awayred',
    'homeyellow', 'awayyellow'
]

df = df.drop(columns=cols_to_drop)

df.head()
```

```python
features_to_roll = [
    'ftgoalsdiff', 'htgoalsdiff', 'shotsdiff', 'targetdiff',
    'foulsdiff', 'cornersdiff', 'yellowdiff', 'reddiff'
]

home_stats = df[['matchdate', 'hometeam'] + features_to_roll].copy()
home_stats['match_index'] = home_stats.index
home_stats = home_stats.rename(columns={'hometeam': 'team'})

away_stats = df[['matchdate', 'awayteam'] + features_to_roll].copy()
away_stats['match_index'] = away_stats.index
away_stats = away_stats.rename(columns={'awayteam': 'team'})

for col in features_to_roll:
    away_stats[col] = -away_stats[col]

team_stats = pd.concat([home_stats, away_stats])
team_stats = team_stats.sort_values(['team', 'matchdate'])

rolling_cols = [f'{col}_last5' for col in features_to_roll]
team_stats[rolling_cols] = team_stats.groupby('team')[features_to_roll].transform(
    lambda x: x.shift(1).rolling(window=5).mean()
)

home_rolling = team_stats.loc[home_stats.index, rolling_cols + ['match_index']].set_index('match_index')

stats_home_mapped = team_stats[team_stats['match_index'].isin(home_stats['match_index']) & (team_stats['team'].isin(df['hometeam']))]

home_stats_keys = df[['hometeam']].copy()
home_stats_keys['match_index'] = df.index
home_stats_keys = home_stats_keys.rename(columns={'hometeam': 'team'})

away_stats_keys = df[['awayteam']].copy()
away_stats_keys['match_index'] = df.index
away_stats_keys = away_stats_keys.rename(columns={'awayteam': 'team'})

home_rolling = pd.merge(home_stats_keys, team_stats[['team', 'match_index'] + rolling_cols], on=['team', 'match_index'], how='left').set_index('match_index')
away_rolling = pd.merge(away_stats_keys, team_stats[['team', 'match_index'] + rolling_cols], on=['team', 'match_index'], how='left').set_index('match_index')

for col in features_to_roll:
    df[f'{col}last5'] = home_rolling[f'{col}_last5'] - away_rolling[f'{col}_last5']

df = df.drop(columns=features_to_roll)

print(f"Shape before dropping NaNs: {df.shape}")
df = df.dropna()
print(f"Shape after dropping NaNs: {df.shape}")

df.head()
```

```python
df["homeelo"] = df["homeelo"].fillna(
    df["hometeam"].map(team_mean)
)

df["homeelo"] = df["homeelo"].fillna(global_mean)

df["awayelo"] = df["awayelo"].fillna(
    df["awayteam"].map(team_mean)
)

df["awayelo"] = df["awayelo"].fillna(global_mean)
```

```python
features = [
    ("homeyellow",  "hometeam"),
    ("awayyellow",  "awayteam"),
    ("homered",     "hometeam"),
    ("awayred",     "awayteam"),
    ("homeshots",   "hometeam"),
    ("awayshots",   "awayteam"),
    ("hometarget",  "hometeam"),
    ("awaytarget",  "awayteam"),
    ("homefouls",   "hometeam"),
    ("awayfouls",   "awayteam"),
    ("homecorners", "hometeam"),
    ("awaycorners", "awayteam")
]

for col, team_col in features:
    team_median = df.groupby(team_col)[col].median()
    global_median = df[col].median()
    
    df[col] = df[col].fillna(df[team_col].map(team_median))
    df[col] = df[col].fillna(global_median)
```

```python
df["season"] = df["matchdate"].dt.year.where(
    df["matchdate"].dt.month >= 7,
    df["matchdate"].dt.year - 1
)

cols = [
    "over25",
    "under25",
    "maxover25",
    "maxunder25",
    "handisize",
    "handihome",
    "handiaway"
]

for col in cols:
    
    group_median = df.groupby(["division","season"])[col].transform("median")
    
    df[col] = df[col].fillna(group_median)
    df[col] = df[col].fillna(df[col].median())
```

```python
numeric_cols = df.select_dtypes(include=['float64','int64','int32']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].median()
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 2.5 * IQR
    upper = Q3 + 2.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    
    if len(outliers) > 0:
        print(f"Feature: {col}")
        print(f"Q1: {Q1}")
        print(f"Q2 (Median): {Q2}")
        print("-" * 30)
```

```python
cols_to_clip = ['oddwin', 'oddloss', 'maxwin', 'maxloss']

for col in cols_to_clip:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    upper = Q3 + 2.5 * IQR
    
    df[col] = df[col].clip(upper=upper)

print("Clipped values to upper bound for:", cols_to_clip)
```

```python
df['elodiff'] = df['homeelo'] - df['awayelo']
df['ftgoalsdiff'] = df['fthome'] - df['ftaway']
df['htgoalsdiff'] = df['hthome'] - df['htaway']
df['shotsdiff'] = df['homeshots'] - df['awayshots']
df['targetdiff'] = df['hometarget'] - df['awaytarget']
df['foulsdiff'] = df['homefouls'] - df['awayfouls']
df['cornersdiff'] = df['homecorners'] - df['awaycorners']
df['reddiff'] = df['homered'] - df['awayred']
df['yellowdiff'] = df['homeyellow'] - df['awayyellow']
df['odddiff'] = df['oddwin'] - df['oddloss']
df['maxodddif'] = df['maxwin'] - df['maxloss']
df['diff25'] = df['over25'] - df['under25']
df['diffmax25'] = df['maxover25'] - df['maxunder25']
df['handidiff'] = df['handihome'] - df['handiaway']

cols_to_drop = [
    'homeelo', 'awayelo',
    'form5home', 'form5away',
    'fthome', 'ftaway',
    'hthome', 'htaway',
    'homeshots', 'awayshots',
    'hometarget', 'awaytarget',
    'homefouls', 'awayfouls',
    'homecorners', 'awaycorners',
    'homered', 'awayred',
    'homeyellow', 'awayyellow',
    'oddwin', 'oddloss',
    'maxwin', 'maxloss',
    'over25', 'under25',
    'maxover25', 'maxunder25',
    'handiaway', 'handihome'
 ]

df = df.drop(columns=cols_to_drop)

df.head()
```

```python
features_to_roll = [
    'ftgoalsdiff', 'htgoalsdiff', 'shotsdiff', 'targetdiff',
    'foulsdiff', 'cornersdiff', 'yellowdiff', 'reddiff', 'htresult',
    'odddiff', 'maxodddif', 'diff25', 'diffmax25', 'handidiff', 'handisize'
]

home_stats = df[['matchdate', 'hometeam'] + features_to_roll].copy()
home_stats['match_index'] = home_stats.index
home_stats = home_stats.rename(columns={'hometeam': 'team'})

away_stats = df[['matchdate', 'awayteam'] + features_to_roll].copy()
away_stats['match_index'] = away_stats.index
away_stats = away_stats.rename(columns={'awayteam': 'team'})

for col in features_to_roll:
    away_stats[col] = -away_stats[col]

team_stats = pd.concat([home_stats, away_stats])
team_stats = team_stats.sort_values(['team', 'matchdate'])


rolling_cols = [f'{col}_last5' for col in features_to_roll]
team_stats[rolling_cols] = team_stats.groupby('team')[features_to_roll].transform(
    lambda x: x.shift(1).rolling(window=5).mean()
)

home_stats_keys = df[['hometeam']].copy()
home_stats_keys['match_index'] = df.index
home_stats_keys = home_stats_keys.rename(columns={'hometeam': 'team'})

away_stats_keys = df[['awayteam']].copy()
away_stats_keys['match_index'] = df.index
away_stats_keys = away_stats_keys.rename(columns={'awayteam': 'team'})

home_rolling = pd.merge(home_stats_keys, team_stats[['team', 'match_index'] + rolling_cols], on=['team', 'match_index'], how='left').set_index('match_index')
away_rolling = pd.merge(away_stats_keys, team_stats[['team', 'match_index'] + rolling_cols], on=['team', 'match_index'], how='left').set_index('match_index')

for col in features_to_roll:
    df[f'{col}last5'] = home_rolling[f'{col}_last5'] - away_rolling[f'{col}_last5']

df = df.drop(columns=features_to_roll)

print(f"Shape before dropping NaNs: {df.shape}")
df = df.dropna()
print(f"Shape after dropping NaNs: {df.shape}")

df.head()
```

```python
train = df[df['matchdate'].dt.year.isin([2020, 2021, 2022])]
val = df[df['matchdate'].dt.year == 2023]
test = df[df['matchdate'].dt.year.isin([2024, 2025])]

print(f"Train set shape: {train.shape}")
print(f"Validation set shape: {val.shape}")
print(f"Test set shape: {test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


predictors = train.columns.drop(['ftresult','division', 'matchdate', 'hometeam', 'awayteam', 'season'])
target = 'ftresult'

lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(train[predictors], train[target])

y_pred_val = lr_model.predict(val[predictors])

accuracy = accuracy_score(val[target], y_pred_val)
print(f"Validation Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(val[target], y_pred_val))

cm = confusion_matrix(val[target], y_pred_val)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Validation Set')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42, n_estimators=1000)
rf_model.fit(train[predictors], train[target])

y_pred_val_rf = rf_model.predict(val[predictors])

accuracy_rf = accuracy_score(val[target], y_pred_val_rf)
print(f"Random Forest Validation Accuracy: {accuracy_rf:.4f}")
print("\nClassification Report:")
print(classification_report(val[target], y_pred_val_rf))

cm_rf = confusion_matrix(val[target], y_pred_val_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Random Forest (Validation Set)')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(train[predictors], train[target])

y_pred_val_xgb = xgb_model.predict(val[predictors])

accuracy_xgb = accuracy_score(val[target], y_pred_val_xgb)
print(f"XGBoost Validation Accuracy: {accuracy_xgb:.4f}")
print("\nClassification Report:")
print(classification_report(val[target], y_pred_val_xgb))

cm_xgb = confusion_matrix(val[target], y_pred_val_xgb)
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - XGBoost (Validation Set)')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
teams_order = [
    "Barcelona",
    "Atletico",
    "Arsenal",
    "Sporting",
    "PSG",
    "Liverpool",
    "Real Madrid",
    "Bayern",
]


cols_to_remove = ["elodiff"]

features_to_aggregate = [
    "ftgoalsdiff", "htgoalsdiff",
    "shotsdiff", "targetdiff", "foulsdiff",
    "cornersdiff", "yellowdiff", "reddiff", "htresult",
    "odddiff", "maxodddif", "diff25", "diffmax25", "handidiff",
    "handisize",
]

rename_map = {
    "ftgoalsdiff": "ftgoalsdifflast5",
    "htgoalsdiff": "htgoalsdifflast5",
    "shotsdiff": "shotsdifflast5",
    "targetdiff": "targetdifflast5",
    "foulsdiff": "foulsdifflast5",
    "cornersdiff": "cornersdifflast5",
    "yellowdiff": "yellowdifflast5",
    "reddiff": "reddifflast5",
    "htresult": "htresultlast5",
    "odddiff": "odddifflast5",
    "maxodddif": "maxodddiflast5",
    "diff25": "diff25last5",
    "diffmax25": "diffmax25last5",
    "handidiff": "handidifflast5",
}

ucl_df = ucl_df.drop(columns=cols_to_remove, errors="ignore").copy()

group_idx = ucl_df.index.to_series().reset_index(drop=True) // 5

ucl_df = (
    ucl_df[features_to_aggregate]
    .groupby(group_idx)
    .mean()
    .rename(columns=rename_map)
)

ucl_df.insert(0, "Team", teams_order)
ucl_df
```

```python
fixtures = [
    ("Barcelona", "Atletico"),
    ("Arsenal", "Sporting"),
    ("PSG", "Liverpool"),
    ("Real Madrid", "Bayern"),
]

team_features = ucl_df.set_index("Team").copy()
numeric_cols = team_features.select_dtypes(include=[np.number]).columns.tolist()

h2h_rows = []
for team1, team2 in fixtures:
    diff_row = team_features.loc[team1, numeric_cols] - team_features.loc[team2, numeric_cols]

    if "elo" in diff_row.index:
        diff_row = diff_row.rename(index={"elo": "elodiff"})

    row = {
        "team1": team1,
        "team2": team2,
    }
    row.update(diff_row.to_dict())
    h2h_rows.append(row)

h2h_df = pd.DataFrame(h2h_rows)
h2h_df
```

```python
p_draw = 1 / df["odddraw"]
p_away = 1 / df["oddaway"]

df["oddwin"] = df["oddhome"]
df["oddloss"] = 1 / (p_draw + p_away)

p_draw = 1 / df["maxdraw"]
p_away = 1 / df["maxaway"]

df["maxwin"] = df["maxhome"]
df["maxloss"] = 1 / (p_draw + p_away)

df = df.drop(columns=["oddhome","odddraw","oddaway","maxhome","maxdraw","maxaway"])
```

```python
teams = pd.concat([
    df[["hometeam", "homeelo"]].rename(columns={"hometeam":"team", "homeelo":"elo"}),
    df[["awayteam", "awayelo"]].rename(columns={"awayteam":"team", "awayelo":"elo"})
])

teams["is_null"] = teams["elo"].isnull()

summary = teams.groupby("team")["is_null"].value_counts().unstack(fill_value=0)
summary.columns = ["non_null", "null"]
summary.sort_values("null", ascending=False)
```

```python
null_rows = df[df["homeelo"].isnull() | df["awayelo"].isnull()]
pd.concat([null_rows["division"]]).value_counts()
```

```python
teams = pd.concat([
    df[["hometeam", "homeelo"]].rename(columns={"hometeam":"team","homeelo":"elo"}),
    df[["awayteam", "awayelo"]].rename(columns={"awayteam":"team","awayelo":"elo"})
])

team_mean = teams.groupby("team")["elo"].mean()
global_mean = teams["elo"].mean()
```

```python
df["homeelo"] = df["homeelo"].fillna(
    df["hometeam"].map(team_mean)
)

df["homeelo"] = df["homeelo"].fillna(global_mean)

df["awayelo"] = df["awayelo"].fillna(
    df["awayteam"].map(team_mean)
)

df["awayelo"] = df["awayelo"].fillna(global_mean)
```

```python
print(df["homeelo"].isnull().sum())
print(df["awayelo"].isnull().sum())
```

```python
df.info()
```

```python
features = [
    ("homeshots",   "hometeam"),
    ("awayshots",   "awayteam"),
    ("hometarget",  "hometeam"),
    ("awaytarget",  "awayteam"),
    ("homefouls",   "hometeam"),
    ("awayfouls",   "awayteam"),
    ("homecorners", "hometeam"),
    ("awaycorners", "awayteam")
]

for col, team_col in features:
    team_mean = df.groupby(team_col)[col].mean()
    global_mean = df[col].mean()
    
    df[col] = df[col].fillna(df[team_col].map(team_mean))
    df[col] = df[col].fillna(global_mean)

features = [
    ("homeyellow",  "hometeam"),
    ("awayyellow",  "awayteam"),
    ("homered",     "hometeam"),
    ("awayred",     "awayteam")
]

for col, team_col in features:
    team_median = df.groupby(team_col)[col].median()
    global_median = df[col].median()
    
    df[col] = df[col].fillna(df[team_col].map(team_median))
    df[col] = df[col].fillna(global_median)
```

```python
df.info()
```

```python
df["season"] = df["matchdate"].dt.year.where(
    df["matchdate"].dt.month >= 7,
    df["matchdate"].dt.year - 1
)

cols = [
    "over25",
    "under25",
    "maxover25",
    "maxunder25",
    "handisize",
    "handihome",
    "handiaway"
]

for col in cols:
    
    group_median = df.groupby(["division","season"])[col].transform("median")
    
    df[col] = df[col].fillna(group_median)
    df[col] = df[col].fillna(df[col].median())
```

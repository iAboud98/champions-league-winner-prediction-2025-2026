```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "")
df.columns
```

```python
df["matchdate"] = pd.to_datetime(df["matchdate"])
df = df.drop(columns=["matchtime"])
df.head()
```

```python
df = df.drop(columns=["form3home", "form3away"])
```

```python
df[
    ((df["fthome"] > df["ftaway"]) & (df["ftresult"] != "H")) |
    ((df["fthome"] < df["ftaway"]) & (df["ftresult"] != "A")) |
    ((df["fthome"] == df["ftaway"]) & (df["ftresult"] != "D"))
]
mismatches = (
    ((df["fthome"] > df["ftaway"]) & (df["ftresult"] != "H")) |
    ((df["fthome"] < df["ftaway"]) & (df["ftresult"] != "A")) |
    ((df["fthome"] == df["ftaway"]) & (df["ftresult"] != "D"))
)

mismatches.sum()
```

```python
df[
    ((df["hthome"] > df["htaway"]) & (df["htresult"] != "H")) |
    ((df["hthome"] < df["htaway"]) & (df["htresult"] != "A")) |
    ((df["hthome"] == df["htaway"]) & (df["htresult"] != "D"))
]
mismatches = (
    ((df["hthome"] > df["htaway"]) & (df["htresult"] != "H")) |
    ((df["hthome"] < df["htaway"]) & (df["htresult"] != "A")) |
    ((df["hthome"] == df["htaway"]) & (df["htresult"] != "D"))
)

mismatches.sum()
```

```python
df["ftresult"] = (df["ftresult"] == "H").astype(int)
df["htresult"] = (df["htresult"] == "H").astype(int)
```

```python
print(df[df["homeshots"] < 0]["homeshots"].value_counts())
print(df[df["awayshots"] < 0]["awayshots"].value_counts())
```

```python
p_draw = 1 / df["odddraw"]
p_away = 1 / df["oddaway"]

df["oddwin"] = df["oddhome"]
df["oddloss"] = 1 / (p_draw + p_away)

p_draw = 1 / df["maxdraw"]
p_away = 1 / df["maxaway"]

df["maxwin"] = df["maxhome"]
df["maxloss"] = 1 / (p_draw + p_away)

df = df.drop(columns=["oddhome","odddraw","oddaway","maxhome","maxdraw","maxaway"])
```
